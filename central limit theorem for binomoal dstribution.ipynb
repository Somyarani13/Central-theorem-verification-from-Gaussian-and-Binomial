```python
# Importing the libraries
import random
import math
import matplotlib.pyplot as plt
```

```python
p = 0.7
q = 1-p
sample_size = 30
```

```python
main_cumulative_list = []
for i in range(0 , 1000):
    cumulative_prob = []                #this list empties the previous list elememnts for reassigning the next elememts to that list
    for j in range(0, sample_size):
        
        x = random.uniform(0,1)
        cumulative_prob.append(x)
    
    main_cumulative_list.append(cumulative_prob)
    
print(main_cumulative_list)

    
```

```python
def facto(n):
    num = 1
    while n>=1:
        num = num*n
        n = n - 1
    return num

y = facto(6)
print(y)
    
```

```python
k_list = []   #this list contains 1000 lists of length 30 elements each
for i in range(0 , len(main_cumulative_list)):
    k_dummy = []           #loop iteratring for the 1000 lists
    
    for j in range(0, len(main_cumulative_list[i])): #loop iterates through the lists
        
        kcumulative = 0
        k = 0
        

        while main_cumulative_list[i][j] > kcumulative: #this condition checks for the lists and their respective elements
            
            kcumulative = kcumulative + (facto(sample_size) / (facto(k) * facto(sample_size - k))) * (p ** k) * ((1-p)**(sample_size - k)) 
            k = k + 1
        k_dummy.append(k)
        
    k_list.append(k_dummy)
    
print(k_list)

    
    

        
        
```

```python
mean_list = []

for i in range(0, len(k_list)):
    
    total = 0
    
    for j in range(0, len(k_list[i])):
        
        total = total + k_list[i][j]
    
    sample_mean = total/len(k_list[i])
    
    mean_list.append(sample_mean)

print(mean_list)
        
        
        
        
        
```

```python
sample_mean = 0

for i in range(0 , len(mean_list)):
    
    sample_mean = sample_mean + mean_list[i]
    
main_mean = sample_mean/len(mean_list)

print(main_mean)
```

```python
mean_list.sort()
print(mean_list)
```

```python
#unique elements
Unique_elem = []

for i in range(0, len(mean_list)-1):
    
    if mean_list[i] != mean_list[i + 1]:
        
        Unique_elem.append(mean_list[i])
               
print(Unique_elem)
```

```python
#frequency

index = 0

frequency = [1]*len(Unique_elem)

for i in range(0, len(mean_list)-1):
    
    if mean_list[i] == mean_list[i+1]:
        
        frequency[index] = frequency[index] + 1
        
    else:
        
        index = index + 1
        
print(frequency)
        
        
    
    
```
